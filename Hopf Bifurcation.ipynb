#PHASE PORTRAIT

##Phase Portrait 1

This means,
x(t) on x - axis and y(t) on y - axis.

```
fig = plt.figure()

def sistem(x, y):
    # supercritical
    # xdot = miu*x - y - x*(x**2 + y**2)
    # ydot = x + miu*y - y*(x**2 + y**2)
    
    # subkritikal
    # xdot = miu*x - y + x*(x**2 + y**2)
    # ydot = x + miu*y + y*(x**2 + y**2)
    
    # model predator prey
    xdot = r*x*(alpha + x)*(1 - x) - c*x*y
    ydot = -alpha*d*y + (c - d)*x*y
    
    return xdot, ydot

def rungekutta4(x,y):
    j1, k1 = sistem(x, y)
    j2, k2 = sistem(x + j1*h/2, y + k1*h/2)
    j3, k3 = sistem(x + j2*h/2, y + k2*h/2)
    j4, k4 = sistem(x + j3*h, y + k3*h)

    xplus = (j1 + 2*j2 + 2*j3 + j4) * h / 6
    yplus = (k1 + 2*k2 + 2*k3 + k4) * h / 6
    
    return xplus, yplus

def plot_portrait_fase():
    for i in range(len(nilai_awal)):
        x = nilai_awal[i][0]
        y = nilai_awal[i][1]

        xresult = [x]
        yresult = [y]
        for t in range(20000):
            xplus, yplus = rungekutta4(x, y)

            nextx, nexty = x + xplus, y + yplus
            x, y = nextx, nexty
            xresult.append(x)
            yresult.append(y)
        plt.plot(xresult, yresult, color = '#2F2F28')#2F3131 #2F2E33

    # plot equilibrium points
    # hopf superkritikal, subkritikal
    # plt.plot([0], [0], 'ro')
    
    # model predator prey
    x_eq = alpha * d / (c - d)
    y_eq = alpha * r / (c - d) * (1 - alpha * d / (c - d))
    plt.plot([x_eq], [y_eq], 'ro')
    
    # vector field
    # hopf superkritikal, subkritikal
    # X, Y = np.meshgrid(np.linspace(-1.5, 1.5, 20), np.linspace(-1.5, 1.5, 20))
    # U, V = sistem(X, Y)
    
    # model predator prey
    X, Y = np.meshgrid(np.linspace(-0.1, 2, 20), np.linspace(-0.1, 2, 20))
    U, V = sistem(X, Y)
    
    # normalize arrows
    N = np.sqrt(U ** 2 + V ** 2)
    U2, V2 = U/N, V/N
    plt.quiver(X, Y, U2, V2, color = '#31A2AC')
    plt.axis('image')
    
    # plt.axis([-1.5, 1.5, -1.5, 1.5]) # hopf superkritikal, subkritikal
    # plt.title(r"$\mu = $" + str(miu))
    
    plt.axis([-0.1, 1.2, -0.1, 1.0]) # model predator prey
    plt.title(r"$\alpha = $" + str(alpha))
```

```
# Long Term

import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()


def sistem(x, y):
    # supercritical
    # xdot = miu*x - y - x*(x**2 + y**2)
    # ydot = x + miu*y - y*(x**2 + y**2)
    
    # subkritikal
    # xdot = miu*x - y + x*(x**2 + y**2)
    # ydot = x + miu*y + y*(x**2 + y**2)
    
    # model predator prey
    xdot = r * x * (alpha + x) * (1 - x) - c*x*y
    ydot = -alpha * d * y + (c - d) * x * y
    
    return xdot, ydot

def rungekutta4(x,y):
    j1, k1 = sistem(x, y)
    j2, k2 = sistem(x + j1*h/2, y + k1*h/2)
    j3, k3 = sistem(x + j2*h/2, y + k2*h/2)
    j4, k4 = sistem(x + j3*h, y + k3*h)

    xplus = (j1 + 2*j2 + 2*j3 + j4) * h / 6
    yplus = (k1 + 2*k2 + 2*k3 + k4) * h / 6
    
    return xplus, yplus

def plot_portrait_fase():
    for i in range(len(nilai_awal)):
        x = nilai_awal[i][0]
        y = nilai_awal[i][1]

        xresult = [x]
        yresult = [y]
        for t in range(20000): # 200
            xplus, yplus = rungekutta4(x, y)

            nextx, nexty = x + xplus, y + yplus
            x, y = nextx, nexty
            xresult.append(x)
            yresult.append(y)
        time = list(range(20001)) # 201
        plt.plot(time, xresult, color = '#31A2AC')
        plt.plot(time, yresult, color = '#2F2F28')
        
        # plt.xlabel('t')
        # plt.ylabel('x(t) dan y(t)')
        # plt.legend(('x(t)', 'y(t)'), loc='upper right')
        plt.xlabel('waktu')
        plt.ylabel('populasi prey-predator')
        plt.legend(('prey', 'predator'), loc='upper right')

    # plt.title(r"$\mu = $" + str(miu))
    plt.title(r"$\alpha = $" + str(alpha))
```
